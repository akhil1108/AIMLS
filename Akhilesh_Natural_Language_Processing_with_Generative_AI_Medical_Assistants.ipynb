## Define the response generation function

Create a Python function that takes a query and the loaded LLM as input and returns the generated response. This function should handle the model parameters.


```python
# function to generate, process, and return the response from the LLM
def generate_response(user_prompt):

    # System message
    system_message = """
    [INST]<<SYS>> Respond to the user question based on the user prompt<</SYS>>[/INST]
    """

    # Combine user_prompt and system_message to create the prompt
    prompt = f"{user_prompt}\n{system_message}"

    # Generate a response from the LLaMA model
    response = llm(
        prompt=prompt,
        max_tokens=128,
        temperature=0.01,
        top_p=0.95,
        repeat_penalty=1.2,
        top_k=50,
        stop=['INST'],
        echo=False
    )

    # Extract and return the response text
    response_text = response["choices"][0]["text"]
    return response_text
```

Based on the results from the different prompt and parameter combinations:

* **Increased `max_tokens` helps prevent truncated responses:** The responses with `max_tokens=1024` are noticeably longer and more complete than those with `max_tokens=512`. This is important for getting comprehensive answers, especially for complex medical queries.
* **Detailed prompts lead to more structured and focused answers:** Prompts that explicitly define the desired output structure (e.g., using headings or bullet points) and the role of the LLM (e.g., "expert medical professional") tend to produce more organized and relevant responses.
* **Higher temperature can introduce variability but might also lead to less accurate or consistent information:** While a higher temperature might generate more diverse answers, it can also increase the likelihood of less precise or even incorrect medical information. For a medical application, a lower temperature is generally preferable for factual accuracy.
* **The "Very Detailed Prompt" combinations provide the most comprehensive information:** These prompts, especially when combined with a lower temperature, elicit more in-depth answers covering symptoms, diagnostics, treatments, and complications. This aligns with the goal of providing robust quick decision support.
* **Even with detailed prompts, the output can still be inconsistent:** The LLM's responses can vary slightly even with the same prompt and parameters. This highlights the inherent variability of LLMs and the need for potential post-processing or validation of the generated information in a real-world medical application.

```python
# Define the question to be answered
question = "What are the necessary precautions and treatment steps for a person who has fractured their leg during a hiking trip, and what should be considered for their care and recovery?"

# Define different prompt engineering techniques and parameter combinations
prompt_combinations = [
    {
        "name": "Extract Precautions and Treatments",
        "prompt_template": "[INST]<<SYS>> As a medical expert, extract and list the necessary precautions and treatment steps for a fractured leg based on the following query. Focus only on these two aspects.<</SYS>>[/INST]\nUser Query: {user_prompt}",
        "parameters": {
            "max_tokens": 1024,
            "temperature": 0.01,
            "top_p": 0.95,
            "repeat_penalty": 1.2,
            "top_k": 50,
            "stop": []
        }
    },
    {
        "name": "Extract Care and Recovery Considerations",
        "prompt_template": "[INST]<<SYS>> As a medical expert, extract and list the considerations for care and recovery of a fractured leg based on the following query. Focus only on these aspects.<</SYS>>[/INST]\nUser Query: {user_prompt}",
        "parameters": {
            "max_tokens": 1024,
            "temperature": 0.01,
            "top_p": 0.95,
            "repeat_penalty": 1.2,
            "top_k": 50,
            "stop": []
        }
    },
    {
        "name": "Combined Extraction (Precautions, Treatment, Care, Recovery)",
        "prompt_template": "[INST]<<SYS>> As a medical expert, extract and list the necessary precautions, treatment steps, and considerations for care and recovery for a fractured leg based on the following query. Present the information clearly.<</SYS>>[/INST]\nUser Query: {user_prompt}",
        "parameters": {
            "max_tokens": 2048,
            "temperature": 0.01,
            "top_p": 0.95,
            "repeat_penalty": 1.2,
            "top_k": 50,
            "stop": []
        }
    },
     {
        "name": "Combined Extraction + Higher Temperature",
        "prompt_template": "[INST]<<SYS>> As a medical expert, extract and list the necessary precautions, treatment steps, and considerations for care and recovery for a fractured leg based on the following query. Present the information clearly.<</SYS>>[/INST]\nUser Query: {user_prompt}",
        "parameters": {
            "max_tokens": 2048,
            "temperature": 0.7, # Higher temperature
            "top_p": 0.95,
            "repeat_penalty": 1.2,
            "top_k": 50,
            "stop": []
        }
    },
     {
        "name": "Filtered Information on Immediate Steps",
        "prompt_template": "[INST]<<SYS>> As a medical expert, provide only the immediate necessary precautions and treatment steps for a fractured leg sustained during a hiking trip, based on the following query.<</SYS>>[/INST]\nUser Query: {user_prompt}",
        "parameters": {
            "max_tokens": 1024,
            "temperature": 0.01,
            "top_p": 0.95,
            "repeat_penalty": 1.2,
            "top_k": 50,
            "stop": []
        }
    }
]

results = {}

print(f"\n--- Answering Question: {question} ---")
for combination in prompt_combinations:
    combination_name = combination["name"]
    prompt_template = combination["prompt_template"]
    parameters = combination["parameters"]

    results[combination_name] = {}

    print(f"--- Testing Combination: {combination_name} ---")
    prompt = prompt_template.format(user_prompt=question)
    response = llm(prompt=prompt, **parameters)
    response_text = response["choices"][0]["text"]
    results[combination_name][question] = response_text
    print(f"Answer: {response_text}\n")
```

## Observations

Based on the prompt engineering combinations for the fractured leg question:

*   **Extraction prompts work:** Prompts explicitly asking the model to extract specific information (precautions, treatment, care, recovery) are effective.
*   **Filtering is possible:** The model can be instructed to provide only specific types of information (e.g., immediate steps).
*   **Combined instructions help organization:** Asking for multiple extractions and clear presentation leads to more structured answers.
*   **Temperature still affects detail:** Higher temperatures can introduce more varied points, but lower temperatures keep it focused and likely accurate for medical steps.
*   **Prompt clarity is key for extraction:** The more clearly you define what to extract, the better the results.

### System and User Prompt Template

```python
qna_system_message = """You are a helpful, respectful and honest medical assistant. Always answer as helpfully as possible, while being safe. Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature.

If a question does not make any sense, or is not factually coherent, explain why instead of answering something incorrect. If you don't know the answer to a question, please don't share false information. Provide the answer from the medical documents provided and if the answer is not in the documents, state that you don't have the answer.

Here are the medical documents:
{context}"""

qna_user_message_template = """Now, answer the following question using the medical documents provided above:\nQuestion: {question}"""
```

### Response Function

```python
def RAG(user_input, llm, k=10,max_tokens=500,temperature=0.4,top_p=0.95,top_k=25):
    """
    Args:
        user_input: Takes a user input for which the response should be retrieved from the vectorDB.
        llm: The LLM to be used for generating the response
    Returns:
        The generated response based on the user query and the context from the knowledge base
    """
    # Ensure the retriever uses the 'k' passed to the RAG function
    relevant_document_chunks = retriever.get_relevant_documents(user_input, k=k)
    context_list = [d.page_content.replace("\t", " ") for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)



    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""


    # Quering an LLM
    try:
        response = llm(
                prompt=prompt,
                max_tokens=max_tokens,
                temperature=temperature,
                top_p=top_p,
                repeat_penalty=1.2,
                top_k=top_k,
                stop=['INST'],
                echo=False
                )

        prediction =  response["choices"][0]["text"]

    except Exception as e:
        prediction = f'Sorry, I encountered the following error: \n {e}'

    return  prediction
```

## Output Evaluation

```python
groundedness_rater_system_message = """

You will be presented a ###Question, ###Context used by the AI system and AI generated ###Answer.

Your task is to judge the extent to which the ###Answer is derived from ###Context.

Rate it 1 - if The ###Answer is not derived from the ###Context at all
Rate it 2 - if The ###Answer is derived from the ###Context only to a limited extent
Rate it 3 - if The ###Answer is derived from ###Context to a good extent
Rate it 4 - if The ###Answer is derived from ###Context mostly
Rate it 5 - if The ###Answer is is derived from ###Context completely

Please note: Make sure you give a single overall rating in the range of 1 to 5 along with an overall explanation.

"""
```

```python
relevance_rater_system_message = """

You will be presented with a ###Question, the ###Context used by the AI system to generate a response, and the AI-generated ###Answer.

Your task is to judge the extent to which the ###Answer is relevant to the ###Question, considering whether it directly addresses the key aspects of the ###Question based on the provided ###Context.

Rate the relevance as follows:
- Rate 1 – The ###Answer is not relevant to the ###Question at all.
- Rate 2 – The ###Answer is only slightly relevant to the **###Question**, missing key aspects.
- Rate 3 – The ###Answer is moderately relevant, addressing some parts of the **###Question** but leaving out important details.
- Rate 4 – The ###Answer is mostly relevant, covering key aspects but with minor gaps.
- Rate 5 – The ###Answer is fully relevant, directly answering all important aspects of the **###Question** with appropriate details from the **###Context**.

Note: Provide a single overall rating in the range of 1 to 5, along with a brief explanation of why you assigned that score.
"""
```

```python
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}
"""
```

```python
def generate_ground_relevance_response(user_input,k=2,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template

    rag_answer_max_tokens = 500
    answer = RAG(user_input, llm, k=5, max_tokens=rag_answer_max_tokens, temperature=temperature, top_p=top_p, top_k=top_k)

    full_relevant_document_chunks = retriever.get_relevant_documents(query=user_input)
    limited_relevant_document_chunks = full_relevant_document_chunks[:k]

    context_list_for_rater = [d.page_content for d in limited_relevant_document_chunks]
    context_for_rater = ". ".join(context_list_for_rater)

    # Combining user_prompt and system_message to create the prompt for groundedness rater
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_rater, question=user_input, answer=answer)}
                [/INST]"""

    # Combininb user_prompt and system_message to create the prompt for relevance rater
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_rater, question=user_input, answer=answer)}
                [/INST]"""

    # Generating groundedness rating
    try:
        response_1 = llm(
                prompt=groundedness_prompt,
                max_tokens=max_tokens,
                temperature=temperature,
                top_p=top_p,
                top_k=top_k,
                stop=['INST'],
                )
        groundedness_rating = response_1['choices'][0]['text']
    except Exception as e:
        groundedness_rating = f'Sorry, Groundedness rating encountered the following error: \n {e}'

    # Generatinb relevance rating
    try:
        response_2 = llm(
                prompt=relevance_prompt,
                max_tokens=max_tokens,
                temperature=temperature,
                top_p=top_p,
                top_k=top_k,
                stop=['INST'],
                )
        relevance_rating = response_2['choices'][0]['text']
    except Exception as e:
        relevance_rating = f'Sorry, Relevance rating encountered the following error: \n {e}'

    return groundedness_rating,relevance_rating
```

```python
user_input = "What are the first-line options and alternatives for managing rheumatoid arthritis?"
ground,rel = generate_ground_relevance_response(user_input)

print(ground,end="\n\n")
print(rel)
```

#Business Insights and Recommendations


* Vector database creation time increases with the number of pages in the PDF document.
* Retrieval parameter k is critical as the answer can be spread across multiple contexts.
* chunk_overlap ensures coherence, especially when context spans across chunks.
* max_tokens depends on query complexity; higher values yield detailed responses, while simple queries result in concise outputs despite large token limits due to prompt design and zero temperature.
* Refine prompt design and temperature settings to control response length and creativity.
* Continuously adjust RAG parameters based on specific use cases for optimal performance.
* Prioritize groundedness and relevance in evaluations to ensure reliable and contextually accurate outputs.
* Establish a feedback loop to fine-tune parameters, improving performance for diverse query types.
